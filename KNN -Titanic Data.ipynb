```python
import pandas  as pd
import numpy as np
```

```python
#Read the training data and testing data
training_data = pd.read_excel('Titanic_Survival_train.xls', index_col=0, header=0)
testing_data = pd.read_excel('Titanic_Survival_test.xls', index_col=0, header=0)

print(training_data.shape)
print(testing_data.shape)
```

```python
#feature selection 
training_data = training_data[['Name','Pclass','Sex','Age','Survived']]
testing_data = testing_data[['Name','Pclass','Sex','Age','Survived']]
```

```python
#finding the missing values
print(training_data.isnull().sum())
print()
print(testing_data.isnull().sum())
```

```python
#imputing numerical missing data with mean value

training_data["Age"].fillna(training_data["Age"].mean(),inplace=True)
print(training_data.isnull().sum())

#imputing numerical missing data with mean value

testing_data["Age"].fillna(testing_data["Age"].mean(),inplace=True)
print(testing_data.isnull().sum())
```

```python
training_data["sex1"]=training_data["Sex"].map({"male":0,"female":1})
```

```python
#transforming categorical data to numerical

from sklearn import preprocessing

colname=['Sex']

le=preprocessing.LabelEncoder()

for x in colname:
    training_data[x]=le.fit_transform(training_data[x])
    testing_data[x]=le.fit_transform(testing_data[x])
    
# training_data['Sex']=training_data['Sex'].map({"female":0, "male":1})
# testing_data['Sex']=testing_data['Sex'].map({"female":0, "male":1})
```

```python
training_data.head()
```

```python
training_data.drop("sex1", inplace=True, axis=1)
```

```python
X_train=training_data.values[:870,  1:-1] # variables from pclass to age [ observation ,  columns]
Y_train=training_data.values[:870,   -1] # passangersIds from 1 to 870 

# include test data in training 
# X_train=training_data.values[:,  1:-1]
# Y_train=training_data.values[:,   -1] 


Y_train=Y_train.astype(int)

 


X_test=testing_data.values[:, 1:-1]
Y_test=testing_data.values[:, -1]
Y_test=Y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# mean = 0 and std = 1

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scale = scaler.transform(X_train)
X_test_scale = scaler.transform(X_test)


# while working on separate train and test data, fit the scaler oject only uplon the train data
# and use the same scaler object to transform both the train as well as the test data.
```

```python
#predicting using the KNeighbors_Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN = KNeighborsClassifier(n_neighbors=15)


#fit the model on the data and predict the values
model_KNN.fit(X_train,Y_train)

Y_pred=model_KNN.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report

conf_matrix=confusion_matrix(Y_test,Y_pred)
print(conf_matrix)
print()


print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc_score=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc_score)
```

```python


acc_dict = dict()
for K in range(1, 31):
    
    # create object with k value
    model_KNN = KNeighborsClassifier(K,metric='euclidean')
    
    # model fit 
    model_KNN.fit(X_train, Y_train) 
    
    # model prediction
    Y_pred = model_KNN.predict(X_test)
    
    acc_score= round(accuracy_score(Y_test, Y_pred),2 )
    
    # print accuracy score
    print("Accuracy is: {} for K-Value: {}".format(acc_score,K))
    acc_dict[K] = acc_score
```

```python
for k in acc_dict:
    if acc_dict[k] == max(acc_dict.values()):
        print(k,":", acc_dict[k])
```

```python
model = KNeighborsClassifier(n_neighbors=11, metric="euclidean")

# model fit 
model.fit(X_train, Y_train) 

# model prediction
Y_pred = model.predict(X_test)
```

```python
df = pd.read_csv("Advertising.csv", index_col=0)
```

```python
X  = df[['TV', 'radio', 'newspaper']]
Y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
#Split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,
                                                    random_state=10)  
```

```python
#predicting using the KNeighbors_Classifier
from sklearn.neighbors import KNeighborsRegressor
model_KNN = KNeighborsRegressor(n_neighbors=3, metric="euclidean")
#fit the model on the data and predict the values
model_KNN.fit(X_train,Y_train)
Y_pred=model_KNN.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(Y_test,Y_pred)
mse = mean_squared_error(Y_test,Y_pred)
print(r2)
print(mse)
```
